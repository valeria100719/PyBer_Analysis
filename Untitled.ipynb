```python
# Import dependencies.
import matplotlib.pyplot as plt
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.scatter(x_axis, y_axis, s=y_axis,marker="o", color="blue", label='Chicago')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(10, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
y_axis_larger=[]
for data in y_axis:
 y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Create the plot.
plt.scatter(x_axis, y_axis, s=y_axis, marker="o", color="blue")
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(10, 45)
# Create a title.
plt.title("PyBer Fare by Month")

for data in y_axis:
 y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis,marker="o", color="skyblue", label="Boston")
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
ax.ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.ylim(10, 45)
# Create a title.
ax.title("PyBer Fare by Month")
# Add the legend.
ax.legend()

for data in y_axis:
 y_axis_larger.append(data*3) 
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis])
```
